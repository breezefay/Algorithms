## Lights Out Game

https://en.wikipedia.org/wiki/Lights_Out_(game)

http://web.math.sinica.edu.tw/mathmedia/HTMLarticle.jsp?mID=40307

```python
# Brute-force

import math
k = 7
for i in range(int(math.pow(2, k))):
    
    bits = [0] * k
    index = 0
    while i > 0:
        b = i % 2
        bits[index] = b
        i = i // 2
        index += 1
        
    # 
    test = [0] * k
    for index in range(k):
        if bits[index] == 1:
            test[index] = 1 if test[index] == 0 else 0
            test[(index+1)%k] = 1 if test[(index+1)%k] == 0 else 0
            test[(index-1)%k] = 1 if test[(index-1)%k] == 0 else 0
            
    print(bits, test, "HERE" if test == [1]*k else "")
```
